```python
def plot(func_name, de_best, cs_best):
#     # plot ga fHistory
#     ga_fHistory = pd.read_csv("../data/" + func_name + "/genetic_algorithm/" + func_name + "_fHistory.csv", header=None) 
#     ga_plot_fHistory = ga_fHistory.plot(kind='box',xticks=np.arange(0, 10+1, 100),
#                                         figsize=(8,5),fontsize=15)
#     ga_plot_fHistory.set_xlabel("10th generation",fontsize=15)
#     ga_plot_fHistory.set_ylabel("f(x) value",fontsize=15)
#     ga_plot_fHistory.set_title(func_name + " ga",fontsize=15)
#     fig = ga_plot_fHistory.get_figure()
#     fig.savefig(func_name + "_ga_fHistory.png")

#     # plot de fHistory
#     de_fHistory = pd.read_csv("../data/" + func_name + "/differential_evolution/" + func_name + "_strategy" + de_best + "_fHistory.csv", header=None) 
#     de_plot_fHistory = de_fHistory.plot(kind='box',xticks=np.arange(0, 100+1, 10),
#                                         figsize=(8,5),fontsize=15)
#     de_plot_fHistory.set_xlabel("10th generation",fontsize=15)
#     de_plot_fHistory.set_ylabel("f(x) value",fontsize=15)
#     de_plot_fHistory.set_title(func_name + " de " + de_best,fontsize=15)
#     fig = de_plot_fHistory.get_figure()
#     fig.savefig(func_name+"_de_fHistory.png")
    
    # plot pso fHistory
    pso_fHistory = pd.read_csv("../data/" + func_name + "/particle_swarm/" + func_name + "_fHistory.csv", header=None) 
    pso_plot_fHistory = pso_fHistory.plot(kind='line',
                                        figsize=(8,5),fontsize=15)
    pso_plot_fHistory.set_xlabel("100th generation",fontsize=15)
    pso_plot_fHistory.set_ylabel("f(x) value",fontsize=15)
    pso_plot_fHistory.set_title(func_name + " pso",fontsize=15)
    fig = pso_plot_fHistory.get_figure()
    fig.savefig(func_name + "_pso_fHistory.png")
    
    # plot mfo fHistory
    mfo_fHistory = pd.read_csv("../data/" + func_name + "/moth_flame/" + func_name + "_fHistory.csv", header=None) 
    mfo_plot_fHistory = mfo_fHistory.plot(kind='line',xticks=np.arange(0, 1000+1, 100),
                                        figsize=(8,5),fontsize=15)
    mfo_plot_fHistory.set_xlabel("100th generation",fontsize=15)
    mfo_plot_fHistory.set_ylabel("f(x) value",fontsize=15)
    mfo_plot_fHistory.set_title(func_name + " mfo",fontsize=15)
    fig = mfo_plot_fHistory.get_figure()
    fig.savefig(func_name + "_mfo_fHistory.png")
    
    # plot de fHistory
    cs_fHistory = pd.read_csv("../data/" + func_name + "/cuckoo_search/" + func_name + "_strategy" + cs_best + "_fHistory.csv", header=None) 
    cs_plot_fHistory = cs_fHistory.plot(kind='line',xticks=np.arange(0, 2500+1, 500),
                                        figsize=(8,5),fontsize=15)
    cs_plot_fHistory.set_xlabel("500th generation",fontsize=15)
    cs_plot_fHistory.set_ylabel("f(x) value",fontsize=15)
    cs_plot_fHistory.set_title(func_name + " cs " + cs_best,fontsize=15)
    fig = cs_plot_fHistory.get_figure()
    fig.savefig(func_name+"_cs_fHistory.png")
    
    

    # plot fBest
    fbest_data = pd.DataFrame()
    # ga bests
    ga_fbests = pd.read_csv("../data/"+func_name+"/genetic_algorithm/"+func_name+"_fHistory.csv") 
    fbest_data["ga"] = ga_fbests.iloc[:,-1]
    # de bests
    for x in range(1,11):
        de_fbests = pd.read_csv("../data/"+func_name+"/differential_evolution/"+func_name+"_strategy"+str(x)+"_fHistory.csv", header=None) 
        new_col_name = "de" + str(x)
        fbest_data[new_col_name] = de_fbests.iloc[:,-1]

    # pso bests
    pso_fbests = pd.read_csv("../data/"+func_name+"/particle_swarm/"+func_name+"_fHistory.csv") 
    fbest_data["pso"] = pso_fbests.iloc[:,-1]
    # mfo bests
    mfo_fbests = pd.read_csv("../data/"+func_name+"/moth_flame/"+func_name+"_fHistory.csv") 
    fbest_data["mfo"] = mfo_fbests.iloc[:,-1]
    # cs bests
    for x in range(1,11):
        cs_fbests = pd.read_csv("../data/"+func_name+"/cuckoo_search/"+func_name+"_strategy"+str(x)+"_fHistory.csv", header=None) 
        new_col_name = "cs" + str(x)
        fbest_data[new_col_name] = cs_fbests.iloc[:,-1]
    
    
    fbest_plot = fbest_data.plot(kind='box',
                                        figsize=(10,5),fontsize=15)
    fbest_plot.set_xlabel("algorithm",fontsize=15)
    fbest_plot.set_ylabel("f(x) value",fontsize=15)
    fbest_plot.set_title("f(x) value",fontsize=15)

    fig = fbest_plot.get_figure()
    fig.savefig(func_name+"fBests.png")

    
    # plot time
    ga_time = pd.read_csv("../data/"+func_name+"/genetic_algorithm/"+func_name+"_timeHistory.csv", header=None) 
    de_time = pd.read_csv("../data/"+func_name+"/differential_evolution/"+func_name+"_timeHistory.csv", header=None) 
    pso_time = pd.read_csv("../data/"+func_name+"/particle_swarm/"+func_name+"_timeHistory.csv", header=None) 
    mfo_time = pd.read_csv("../data/"+func_name+"/moth_flame/"+func_name+"_timeHistory.csv", header=None) 
    cs_time = pd.read_csv("../data/"+func_name+"/cuckoo_search/"+func_name+"_timeHistory.csv", header=None) 
    
#     time_data = de_time.transpose();
#     time_data.columns=["de1","de2","de3","de4","de5","de6","de7","de8","de9","de10"]
#     time_data["ga"] = ga_time
#     cols = time_data.columns.tolist()
#     cols = cols[-1:] + cols[:-1]
#     time_data = time_data[cols]
#     time_data[]
    de_time = de_time.transpose();
    cs_time = cs_time.transpose();
    time_data = pd.concat([ga_time, de_time, pso_time, mfo_time, cs_time], axis=1, sort=False)
    
    time_plot = time_data.plot(kind='box',
                                        figsize=(10,5),fontsize=15)
    time_plot.set_xlabel("algorithm",fontsize=15)
    time_plot.set_ylabel("time",fontsize=15)
    time_plot.set_title("function running time",fontsize=15)

    fig = time_plot.get_figure()
    fig.savefig(func_name+"_time.png")
```
